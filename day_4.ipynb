```python
with open("input4.txt", "r") as f:
    data = f.read().split("\n\n")
```

```python
correct = 0
for i, row in enumerate(data):
    fields = ["byr:", "iyr:", "eyr:", "hgt:", "hcl:", "ecl:", "pid:"]
    if all([f in row for f in fields]):
        correct +=1

print(correct)
```

```python
import re

correct = 0
for i, row in enumerate(data):
    try:
        byr = int(re.search("byr:[0-9]{4}(\s|$)", row).group()[4:])
        iyr = int(re.search("iyr:[0-9]{4}(\s|$)", row).group()[4:])
        eyr = int(re.search("eyr:[0-9]{4}(\s|$)", row).group()[4:])
        hgt = re.search("hgt:[0-9]{2,3}(cm|in)(\s|$)", row).group()[4:]
        hgt_small = 150 if "cm" in hgt else 59
        hgt_large = 193 if "cm" in hgt else 76
        hgt = int(re.search("[0-9]*", hgt).group())
        hcl = re.search("hcl:#[a-f0-9]{6}(\s|$)", row).group()
        ecl = re.search("ecl:(amb|blu|brn|gry|grn|hzl|oth)(\s|$)", row).group()
        pid = re.search("pid:[0-9]{9}(\s|$)", row).group()
        print(byr, iyr, eyr, hgt_small, hgt_large, hgt, hcl, ecl, pid)
        if 1920 <= byr <= 2002 \
            and 2010 <= iyr <= 2020 \
            and 2020 <= eyr <= 2030 \
            and hgt_small <= hgt <= hgt_large:
            correct +=1
    except:
        pass

print(correct)
```
